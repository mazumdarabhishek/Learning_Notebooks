## Question answering models using transformers. 

The question answering use case are one of the most important use cases in current trends. There are two scenarios with in this domain. 
- **ODQA:** Stands for Open Domain Question Answering frameworks. Here the model has access to a external source of information which it can utilize to answer questions. For example - general purpose chat-bots like siri, cortana etc
- **CDQA:** Stands for Closed domain QnA models where the model is supposed to only find answers within a given contextual information. 

Within ODQA (open-domain question answering), we have three model types to be considered:

1. *Retriever* - It is responsible for searching relevant content and retrieve it from knowledge base, according to a question
2. *Reader* - It takes in the question as well as relevant data extracted by retriever, finds out the relevant chunk of information within the data to returns an output. This model alone, can be a reading comprehension model if there was no retriever input. 
3. *Generator* -  It takes questions and if available, get the data from the retriever as well. Followed by return an answer as an output. Reader models can be replaced with generator models. Also, often times generator models are stand alone, meaning it doesn't take any context from the retriever. 

***These three models, when applied to ODQA can appear in three different architectures:***

- Retriever-Reader (open-book ODQA)

- Retriever-Generator (open-book ODQA)

- Generator (closed-book ODQA)
